```python
import pandas as pd

file_path = r"C:/Users/Vahe/Desktop/crypto_analysis.csv"
df = pd.read_csv(file_path)
df.head(20)
```

```python
import matplotlib.pyplot as plt


df['Date'] = pd.to_datetime(df['Date'])

df['Month'] = df['Date'].dt.month

# Plot
plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['High'], marker='o', color='blue', linestyle='-')
plt.title('High values by Month')
plt.xlabel('Month')
plt.ylabel('High')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.tight_layout()
plt.show()
```
